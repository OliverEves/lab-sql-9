```python
#Instructions

import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.metrics import classification_report, f1_score, cohen_kappa_score

import getpass  # To get the password without showing the input
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
#How many distinct (different) actors' last names are there?

data = pd.read_sql_query('SELECT COUNT(DISTINCT actor_id) AS NO_of_Actors FROM actor', engine) #only access
data.head()
```

```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

data = pd.read_sql_query('SELECT *, CASE WHEN weekday(rental_date) < 6 THEN "workday" ELSE "weekend" END AS "day_type" FROM rental', engine)
data.tail(20)
```

```python
#Get all films with ARMAGEDDON in the title.

data = pd.read_sql_query('SELECT * FROM film WHERE title LIKE "%%ARMAGEDDON%%"'
, engine)

data.head()
```

```python
#Get 10 the longest films.

data = pd.read_sql_query('SELECT title, length FROM film ORDER BY length DESC LIMIT 10'                        
, engine)
data.head()
```

```python
#How many films include Behind the Scenes content?

data = pd.read_sql_query('SELECT count(special_features) AS " w/Special Features" FROM film WHERE special_features LIKE "%%Behind the Scenes%%"', engine)
data.head()
```

```python
#Which kind of movies (rating) have a mean duration of more than two hours?

data = pd.read_sql_query('SELECT rating AS "Rating Type", round(avg(length),2) AS "Avg. Length" FROM sakila.film GROUP BY rating HAVING round(avg(length),2) > 120'
, engine)

data.head()
```

```python
#Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.

data = pd.read_sql_query('SELECT title AS "Title", length AS "Length", rank() OVER (ORDER BY length) AS Ranking FROM sakila.film WHERE length > 0 AND NOT ISNULL(length)', engine)

data.head()
```
